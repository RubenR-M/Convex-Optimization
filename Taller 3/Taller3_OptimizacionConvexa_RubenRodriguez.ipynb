```python
import matplotlib.pyplot as plt
import numpy as np
```

# Punto 1

```python
equations = ["8-4*x",'5-x','20/7 - x*2/7']

x = np.linspace(-0.0001,10, 100)

fig = plt.figure(figsize = (10, 5))

X,Y = np.meshgrid(x, x)
z = eval("150*X + 300*Y")
plt.pcolor(X,Y,z, cmap="Spectral")

for equation in equations:
  y = eval(equation)
  plt.plot(x, y)
  plt.fill_between(x, -0.97, y, facecolor='white')

plt.legend(equations)

plt.xlim([0, 10])
plt.ylim([0, 10])

plt.show()
```

```python
equations = {"8-4*xx -yx":([0,10],[0, 10]),'5-xx-yx':([0,10],[0,10]),'20/7 - xx*2/7-yx':([0,10],[0,10])}

fig = plt.figure(figsize = (10, 5))
plt.grid(visible=True, which='major', axis='both')

equa = ["8-4*x",'5-x','20/7 - x*2/7']
i=0

for equation in equations:
  x = np.linspace(equations[equation][0][0],equations[equation][0][1], 2)
  y = np.linspace(equations[equation][1][0],equations[equation][1][1], 2)#eval(equation)
  
  xx,yx = np.meshgrid(x, y)
  # Create the plot
  z = eval(equation)
  
  plt.contourf(x, y, z, 100)
  y= eval(equa[i])
  plt.plot(x, y)
  if i<= 3:
    i+=1

plt.legend(equations)
plt.xlim([0, 10])
plt.ylim([0, 10])

plt.show()
```

# Punto 2

```python
equations = ["8-2*x",'6-x']

x = np.linspace(-0.0001,9, 100)

fig = plt.figure(figsize = (10, 5))

X,Y = np.meshgrid(x, x)
z = eval("30000*X + 10000*Y")
plt.pcolor(X,Y,z, cmap="Spectral")

for equation in equations:
  y = eval(equation)
  plt.plot(x, y)
  plt.fill_between(x, y, 9, facecolor='white')

plt.legend(equations)

plt.xlim([0, 7])
plt.ylim([0, 9])

plt.show()
```
